# Tutorial: Jupyter Notebook Terminal Emulator

Audience:
- 想在 Jupyter Notebook 或 JupyterLab 里直接使用类真实终端的人。

Prerequisites:
- 当前环境是 Linux 或 macOS 这类 POSIX 系统。
- Notebook 前端可以访问 CDN 以加载 `xterm.js`。
- 当前目录里已经有 `jupyter_terminal.py`。

Learning goals:
- 直接在 notebook 里启动一个可交互的 shell。
- 使用 `Ctrl+C`、Tab 补全、颜色输出、`vim` / `top` / `htop` 这类 TTY 程序。
- 了解常用按钮和几个实际限制。


## 快速开始

按顺序运行下面两个代码单元。第二个单元执行后，页面里会出现一个真正由 PTY 驱动的终端。

常用操作：
- 直接输入 shell 命令并回车
- `Ctrl+C` 中断当前前台进程
- `Interrupt` 按钮发送 `SIGINT`
- `Restart Shell` 重启 shell 会话
- `Clear` 清屏
- `Fit` 按当前输出区域重新计算行列数


```python
from pathlib import Path
import sys

repo_dir = Path.cwd().resolve()
if str(repo_dir) not in sys.path:
    sys.path.insert(0, str(repo_dir))

from jupyter_terminal import JupyterTerminal

repo_dir
```

```python
term = JupyterTerminal(height=520)
term.display();
```

## 可选：用自定义工作目录或环境变量启动

如果你想让终端从别的目录启动，或者给 shell 增加环境变量，运行下面这个单元。它会新开一个终端实例。


```python
from pathlib import Path

custom_term = JupyterTerminal(
    cwd=str(Path.home()),
    height=420,
    env={
        "TERM_PROGRAM": "jupyter-terminal",
    },
)
custom_term.display();
```

## 说明

这版终端和普通 `ipywidgets.Text` 模拟器的差别在于：

- 后端是 PTY，不是简单的 `subprocess.communicate()`
- 前端用的是 `xterm.js`
- shell 自己处理历史、补全、ANSI 控制序列和全屏程序
- resize 事件会回传到 PTY，因此 `stty size`、`vim`、`less` 这类程序能感知窗口大小变化

已知限制：
- 当前后端只支持 POSIX，不支持 Windows `ConPTY`
- 前端默认从 CDN 加载 `xterm.js`，离线环境需要改成本地静态资源
